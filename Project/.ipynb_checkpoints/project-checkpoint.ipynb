```python
import csv
from pylab import *
f = open('brain.csv', 'rb')
reader = csv.reader(f)
hc=[]
iq=[]
bv=[]
mf=[]

for row in reader :
    a=row[0].replace("\t"," ")
    a=a.split()
    b=a
    b=map(float,a)
    b2=map(int,b)
    hc.append(b[2])
    iq.append(b2[1])
    bv.append(b2[7])
    mf.append(b2[5])
number=0
number1=0
for x in iq[10:]:
    number1+=1
    number=number+x
    avm=float(number)/number1
print avm
number2=0
number3=0
for x in iq[0:10]:
    number3+=1
    number2=number2+x
    avf=float(number2)/number3

print avf
avs=[avm,avf]
print avs



    
why = plt.figure()
why2 = why.add_subplot(111)
why2.set_title('Relationship Between IQ and head Size')
why2.set_ylabel('Full-Scale IQ')
why2.set_xlabel('Head Circumference (cm)')
why2.scatter(hc,iq)
why.savefig("brain.png")
f.close()
```

<b>
<font color="blue" size="4">There is no relationship between the iq and the Head Circumference (cm)<font/><b/>

```python
why3 = plt.figure()
why4 = why3.add_subplot(111)
why4.set_title('Relationship Between IQ and Total Brain Volume')
why4.set_ylabel('Full-Scale IQ')
why4.set_xlabel('Total Brain Volume (cm3)')
why4.scatter(bv,iq)
why3.savefig("brain2.png")
```

<font color="blue" size="4"><b>
There is no relationship between the iq and the brain volume
<font/><b/>

```python
av = plt.figure()
av1 = av.add_subplot(111)
av1.set_title('Relationship Between IQ Gender')
av1.set_ylabel('Full-Scale IQ')
opt=['','', 'Males','','','','','Females']
av1.set_xticklabels(opt)
av1.bar([1,2],avs)
av.savefig("aviqmf.png")
```

<font color="blue" size="4"><b>
